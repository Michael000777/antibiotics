```python
def LMSQ(x,y):
    n   = len(x) # number of values (x)
    sx  = np.sum(x) # sum of x values 
    sy  = np.sum(y) # sum of target values or dependent values (y)
    sxx = np.dot(x,x) # dot product or sum of squared x values 
    sxy = np.dot(x,y) # sum of x y products 
    syy = np.dot(y,y) # sun of squared y values 
    
    denom  = (n*sxx-sx*sx)
    b      = (n*sxy - sx*sy)/denom
    a      = (sy-b*sx)/n
    estim  = np.array([a,b], dtype = float)

    sigma2 = syy + n*a*a + b*b*sxx + 2*a*b*sx - 2*a*sy - 2*b*sxy
    cov    = sigma2 / denom * np.array([[sxx,-sx],[-sx,n]], dtype = float)

    return estim,cov
```

```python

data   = pd.DataFrame()
growth = pd.DataFrame()
count  = {}

for datafile in glob.glob('../ExperimentalData/KillCurves/ctx?.data'):
    x = np.genfromtxt(datafile, delimiter = '\t')
    for row in x:
        if not row[0] in count.keys(): count[row[0]] = 1 # row[0] is a unique idenifier
        else:                          count[row[0]] += 1 # ultimately counts the number of times we process a sample or concentration that could appear in multiple files 
        
        t       = np.arange(len(row[1:])) / 6.
        fit,cov = LMSQ(t,np.log(row[1:]))
        
        data[(row[0],count[row[0]])]   = row[1:]
        growth[(row[0],count[row[0]])] = np.concatenate([fit, cov.flatten()[[0,3,1]]])
        
growth.index = ['logN0', 'alpha', 'Var[logN0]', 'Var[alpha]', 'Cov[logN0,alpha]']
```

```python

data_dict = {}
growth_dict = {}
count = {}

for datafile in glob.glob('../ExperimentalData/KillCurves/ctx?.data'):
    
    x = np.genfromtxt(datafile, delimiter = '\t')
    for row in x:
        if not row[0] in count.keys(): count[row[0]] = 1 # row[0] is a unique idenifier
        else:                          count[row[0]] += 1

        t = np.arange(len(row[1:])) / 6.
        fit,cov = LMSQ(t,np.log(row[1:]))

        key = (row[0], count[row[0]])
        data_dict[key] = row[1:]
        growth_dict[key] = np.concatenate([fit, cov.flatten()[[0, 3, 1]]])

data = pd.DataFrame.from_dict(data_dict, orient='index').T
growth = pd.DataFrame.from_dict(growth_dict, orient='index').T 

growth.index = ['logN0', 'alpha', 'Var[logN0]', 'Var[alpha]', 'Cov[logN0,alpha]']

data
```

```python
def HillFunction(x, alpha0, logmu, gamma, kappa):
    bmk = np.exp(kappa * (x-logmu))
    return alpha0 * (1 - bmk)/(1 + bmk/gamma)

# kappa here represents a parameter that controls the rate of change 
# log of a midpoint conc where curve shifts 
# alpha0 is the maximum asymtote the function can approach
# gamma adjusts the steepness of the hill function
# kappa controls how quickly the function decays or rises.

```

```python
p0 = [1,np.log(1e-2), 2, 5] #these values are the initial  guesses of params for the func; otherwise 1
b0 = []
gr = []

for c in growth.columns:
    if c[0] < threshold and c[0] > 0: # grabbing concs between 0 and 1
        b0.append(c[0])
        gr.append(growth[c]['alpha'])

L0       = np.log(np.array(b0)) # log transformation of the concs
gr       = np.array(gr)

fit, cov = curve_fit(HillFunction, L0, gr, p0 = p0) # cruve fit tries to estimate the params given the observed data

# output
print('alpha0 = {}\nmu     = {}\ngamma  = {}\nkappa  = {}'.format(fit[0],np.exp(fit[1]), fit[2], fit[3]))


# checking results
fig,ax = plt.subplots(1,1,figsize = (15,8))
x      = np.linspace(-7,0,100)

ax.scatter(L0, gr, label = 'data points')
ax.plot(x,HillFunction(x,fit[0],fit[1],fit[2],fit[3]), label = 'final fit', c = 'orange', lw = 3)
ax.plot(x,HillFunction(x,p0[0],p0[1],p0[2],p0[3]), label = 'starting values', c = 'green', lw = 3)

ax.set_ylim([-3,3])
ax.legend()
```
